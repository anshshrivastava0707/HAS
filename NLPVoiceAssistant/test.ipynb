```python
lemmer = nltk.stem.WordNetLemmatizer()
def lemmatiseTokens(tokens):
    for token in tokens:
        return lemmer.lemmatize(token)
```

```python
remove_punctuation = dict((ord(punct),None) for punct in string.punctuation)
print(remove_punctuation)
```

```python
def normalise(text):
    text.lower.translate(remove_punctuation)
    return lemmatiseTokens(nltk.word_tokenize(text.lower.translate(remove_punctuation)))
```

```python
def response(user_response):
    response = ""
    sentTokens.append(user_response)
    TfidfVec = TfidfVectorizer(tokenizer=lemmatiseTokens,stop_words = 'english')
    tfidf = TfidfVec.fit_transform(sentTokens)
    print("tfidf",tfidf,"tfidf[-1]",tfidf[-1])
    
    
    cosine_sim = cosine_similarity(tfidf[-1],tfidf[1])
    print("cosine_sim",cosine_sim)
    idx = cosine_sim.argsort()[0][-1]
    print("whatever this is",idx)
    flat = cosine_sim.flatten()
    flat.sort()
    print("flatboi",flat)
    req_tfidf = flat[0]
    print(req_tfidf)
    
    if(req_tfidf==0):
        response=response+"I am sorry! I don't understand you"
        return response
    else:
        response = response+sentTokens[idx]
        return response

```
